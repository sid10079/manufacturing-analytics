```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sqlalchemy import create_engine

# Connect to MySQL - REPLACE with your actual password
engine = create_engine('mysql+pymysql://root:NewPassword123!@127.0.0.1/manufacturing_analytics')

# Load data from MySQL
df = pd.read_sql('SELECT * FROM machine_data', con=engine)

# Quick look at the data
print(df.shape)
print(df.head())
```

```python
# Check data types and missing values
print(df.info())
print("\n")
print(df.describe())
```

```python
# Missing values
print("Missing values:")
print(df.isnull().sum())

# Failure rate
print("\nMachine Failure Rate:")
print(df['Machine_failure'].value_counts())
print(f"\nFailure %: {df['Machine_failure'].mean()*100:.2f}%")
```

```python
# Plot failure distribution
plt.figure(figsize=(8,5))
sns.countplot(x='Machine_failure', data=df, palette='Set2')
plt.title('Machine Failure Distribution')
plt.xlabel('Machine Failure (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 3, figsize=(15, 8))

sns.histplot(df['Air_temperature_K'], ax=axes[0,0], color='blue')
axes[0,0].set_title('Air Temperature')

sns.histplot(df['Process_temperature_K'], ax=axes[0,1], color='orange')
axes[0,1].set_title('Process Temperature')

sns.histplot(df['Rotational_speed_rpm'], ax=axes[0,2], color='green')
axes[0,2].set_title('Rotational Speed')

sns.histplot(df['Torque_Nm'], ax=axes[1,0], color='red')
axes[1,0].set_title('Torque')

sns.histplot(df['Tool_wear_min'], ax=axes[1,1], color='purple')
axes[1,1].set_title('Tool Wear')

sns.histplot(df['Machine_failure'], ax=axes[1,2], color='brown')
axes[1,2].set_title('Machine Failure')

plt.suptitle('Sensor Reading Distributions', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Between All Variables')
plt.tight_layout()
plt.show()
```

```python
# Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_df.corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Between All Variables')
plt.tight_layout()
plt.show()
```

```python
# Boxplots to compare failed vs non-failed machines
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

sns.boxplot(x='Machine_failure', y='Torque_Nm', data=df, ax=axes[0], palette='Set2')
axes[0].set_title('Torque vs Failure')

sns.boxplot(x='Machine_failure', y='Tool_wear_min', data=df, ax=axes[1], palette='Set2')
axes[1].set_title('Tool Wear vs Failure')

sns.boxplot(x='Machine_failure', y='Rotational_speed_rpm', data=df, ax=axes[2], palette='Set2')
axes[2].set_title('Rotational Speed vs Failure')

plt.suptitle('Sensor Readings vs Machine Failure', fontsize=16)
plt.tight_layout()
plt.show()
```
